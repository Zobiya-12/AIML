# Train test Split

```python
df_feat=df.drop('Cancer',axis=1)#Define a dataframe with only features
df_feat.head()
```

```python
df_target=df['Cancer']#Define a dataframe for cancer detection
df_target.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(df_feat, df_target,test_size=0.20,
                 random_state=101)
```

```python
df_feat.shape, X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

# Predictions and evaluations

```python
predictions=model.predict(X_test)#X-Columns,Y-rows
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
print('Misclassification error rate:',round(np.mean(predictions!=y_test),3))
```

```python
print("Test Data Misclassification error rate:",
      round(np.mean(predictions!=y_test),3))

train_pred = model.predict(X_train)
print("Train Data Misclassification error rate:",
      round(np.mean(train_pred!=y_train),3))
```
