Visualization of Data suing matplotlib

```python
#1. Generate a Data set for Health Care
import matplotlib.pyplot as plt
people = ["kiran", "arun", "vijay", "varun"]
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
#2. Generating a Bar Graph using the above data
plt.figure(figsize=(10, 5))
plt.bar(people, height, color= 'Blue')
plt.xlabel("people")
plt.ylabel("Height (cm)")
plt.title("Height of Individuals")
plt.show()
plt.figure(figsize=(10, 5))
plt.bar(people, weight, color="red")
plt.xlabel("people")
plt.ylabel("Weight (kg)")
plt.title("Weight of Individuals")
plt.show()
plt.figure(figsize=(10, 5))
plt.bar(people, age, color="green")
plt.xlabel("people")
plt.ylabel("age")
plt.title("People and their Age")
plt.show()
# Generating a Scatter Plot using the above data
plt.figure(figsize=(10, 5))
plt.scatter(age, height, c="red")
plt.xlabel("age")
plt.ylabel("Height (cm)")
plt.title("Age vs Height")
plt.show()
plt.figure(figsize=(10,5))
plt.scatter(age, weight, c="blue")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("Age vs Weight")
plt.show()
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.hist(age, edgecolor="red", color="blue", bins=4)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("Histogram of Age")
plt.show()
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.hist(height, bins=4, edgecolor="blue", color="red")
plt.xlabel("height")
plt.ylabel("frequency")
plt.title("Histogram of height")
plt.show()
plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.hist(weight, bins=4, edgecolor="red", color="green")
plt.xlabel("weight")
plt.ylabel("frequency")
plt.title("Histogram of weight")
plt.show()
```
